```python
#폴더별로 파일 이미지 증강하고 자동생성되는 output내 파일 폴더내로 이동하고 output폴더 삭제하기
import Augmentor #이미지 증강 # !pip install Augmentor
import shutil    #파일 옮기기

fdir='/content/drive/MyDrive/인공지능실습/한경/image/'
folderList=os.listdir(fdir)
for folder in folderList:
  img=Augmentor.Pipeline(fdir+folder)   #증강시킬 이미지 폴더 경로
  img.flip_left_right(probability=1.0)  ## 좌우 반전
  img.flip_top_bottom(probability=1.0)  ## 상하 반전
  img.random_distortion(probability=1, grid_width=10, grid_height=10, magnitude=8) ## 왜곡
  img.sample(20)                        ## 증강 이미지 수
  auList=os.listdir(fdir+folder+'/output') #이미지 증강 후 자동으로 생성되는 output폴더 내 파일리스트
  for au in auList:                       #output폴더 내 파일
    shutil.move(fdir+folder+'/output/'+au, fdir+folder+'/'+au) #output폴더 내 파일을 상위 폴더로 옮기기 
  os.rmdir(fdir+folder+"/output") #output폴더 삭제( os.rmdir는 빈 디렉토리일때만 삭제 가능함.)
```

```python
#폴더에 있는 이미지파일을 넘파이배열로 만들고, 폴더명을 레이블로 제작
X=[]; y=[]; label=0
fdir='/content/drive/MyDrive/인공지능실습/한경/image/'
folderList=os.listdir(fdir)
for folder in folderList: #폴더리스트: 진, 슈가, 숫자
  fileList=os.listdir(folder) #폴더 내 파일들 목록
  for item in fileList: #파일리스트에서 개별 이미지파일을 item으로 가져와라.
    img=Image.open(fdir+folder+'/'+item) #이미지 열기
    img=img.resize((50,50)) #이미지 리사이즈
    img=np.array(img) #이미지 배열화
    X.append(img)
    y.append(label)
  label +=1
```

```python
#y라벨 차원 확장 (73,)를 (73,1)로
y=np.reshape(y,(-1,1))
print(np.shape(y))
```

```python
#train test 분할
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.2, random_state=100)
```

```python
# #npz로 저장하기
fdir='/content/drive/MyDrive/인공지능실습/한경/image/'
folderList=os.listdir(fdir) 
np.savez(fdir+'진슈가숫자합', 
         X_train=X_train, y_train=y_train, X_test=X_test, y_test=y_test) #폴더별 이미지파일을 npz로 저장 np.savez(저장경로, 저장데이터)
```
