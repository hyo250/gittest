```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder
from keras.models import Sequential
from keras.layers.core import Dense
```

```python
df=pd.read_csv('./dataset/sonar.csv', header=None)
df.info()
```

```python
np.random.seed(3)
tf.random.set_seed(3)
```

```python
dataset=df.values
X=dataset[:, 0:60]
y_obj=dataset[:,60]
```

```python
#문자열 인코딩
e=LabelEncoder()
e.fit(y_obj)
y=e.transform(y_obj)
```

```python
#모델 설정
model=Sequential()
model.add(Dense(24, input_dim=60, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
X=X.astype('float')
```

```python
#모델 컴파일
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

#모델 실행
model.fit(X, y, epochs=200, batch_size=5)

#결과출력
print('\n 정확도: %4f' %(model.evaluate(X,y)[1]))
```

- 정확도가 1.00이라니... 말도안된... 이게 바로 과적합^^
- 학습셋 테스트셋 구분해서 해보자

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100)
model.fit(X_train, y_train, epochs=130, batch_size=5) #208개 데이터에서 훈련셋이 70%니까 대략 이정도로 epochs한듯.
print('\n 정확도: % 4f'%(model.evaluate(X_test, y_test)[1]))
```

```python
#모델 저장과 재사용
from keras.models import load_model
model.save('sonar_model.h5')
```

# k-fold 교차검증
- 데이터가 적을때. 데이터셋을 증강하는 기술!

```python
from sklearn.model_selection import StratifiedKFold
n_fold=10 #쪼갤 파일 수 
skf=StratifiedKFold(n_splits=n_fold, shuffle=True, random_state=100) #10개의 파일로 쪼개 테스트하는 10겹 교차검증. 
```

```python
acc=[]
for train, test in skf.split(X,y):
  model=Sequential()
  model.add(Dense(24, input_dim=60, activation='relu'))
  model.add(Dense(10, activation='relu'))
  model.add(Dense(1, activation='sigmoid'))
  model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
  model.fit(X[train],y[train], epochs=100, batch_size=5)
  k_acc="%.4f" %(model.evaluate(X[test], y[test])[1])
  acc.append(k_acc)
```

```python

print('% f fold accuracy: ' %n_fold, acc)
```
